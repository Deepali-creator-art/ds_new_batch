```python
from structlinks.DataStructures import Stack

def bfs(graph,start_node):
    s1=Stack.Stack([start_node])
    visited_vertex=set()
    while s1:
        vertex=s1.pop()
        if vertex not in visited_vertex:
            visited_vertex.add(vertex)
            for i in graph[vertex]:
                s1.push(i)
        print(s1)
    return visited_vertex       
graph={'0':['1','2','3'],
       '1':['0','2'],
       '2':['0','1','3'],
       '3':['0','2','4'],
       '4':['2','3','5','6'],
       '5':['4'],
       '6':['4']}
start_node='0'
bfs(graph,start_node)
```

```python
def dfs(graph,start_node):
    s1=[start_node]
    visited_vertex=set()
    while s1:
        vertex=s1.pop()
        if vertex not in visited_vertex:
            visited_vertex.add(vertex)
            print("Visited vertex",visited_vertex)
            s1.extend(graph[vertex]-visited_vertex)
        print("Stack elements")
        print(s1)
    return visited_vertex       
graph={'0':{'1','2','3'},
       '1':{'0','2'},
       '2':{'0','1','3'},
       '3':{'0','2','4'},
       '4':{'2','3','5','6'},
       '5':{'4'},
       '6':{'4'}}
start_node='0'
dfs(graph,start_node)
```
