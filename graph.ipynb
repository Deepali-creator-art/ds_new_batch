```python
def tsp_problem(distance):
    n=len(distance) # n is a no of vertices
     #it is similar to set data type only difference is set is muttable and frozenset is unmuttable
    all_data_points=frozenset(range(n)) #0,1,2,3
    print(all_data_points)
    min_distance=float('inf') # assign each vertex value is an infinity value
    m={} #new set 
    def tsp_calculation(current_vertex,remaining_vertex):
        print("Current vertex",current_vertex,"Remaining vertex",remaining_vertex)
        if not remaining_vertex:
            return distance[current_vertex][0]
        if(current_vertex,remaining_vertex) in m:
         
            return m[(current_vertex,remaining_vertex)]
        
        min_distance=float('inf')
        for i in remaining_vertex:
            print("Iteration :",i)
            new_remaining_vertex=remaining_vertex-{i} #set data uses - minus operator {0,1,2,3}-{0}=>{1,2,3}
            print(new_remaining_vertex)
         
            dist=distance[current_vertex][i]+tsp_calculation(i,new_remaining_vertex)
            print(dist)
            
            min_distance=min(min_distance,dist)
        #return dist
        #print("M=",m)
        m[(current_vertex,remaining_vertex)]=min_distance
        return min_distance
    final_dist=tsp_calculation(0,all_data_points) # call a function
    return final_dist
         
        
distance=[[0,100,15,20],
          [5,0,9,10],
          [6,13,0,12],
          [8,8,9,0]]
result=tsp_problem(distance)
print(result)
```
