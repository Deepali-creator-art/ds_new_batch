```python
#Adcency matrix
G=[[0,1,0,4],[1,0,6,6],[0,6,0,3],[4,6,3,0]]
#variable for infinity
import math
INF=math.inf
#NO.OF VERTEX  
V=len(G)
#NO OF EDGES
no_of_edges=0
selected_vertex=[0,0,0,0]
selected_vertex[0]=True
while(no_of_edges<V-1):
    minimum=INF
    x=0
    y=0
    for i in range(V): #0,1,2
        if selected_vertex[i]:
            for j in range(V): #0,1,2
                if ((not selected_vertex[j]) and G[i][j]):  
                    # not in selected and there is an edge
                    if minimum > G[i][j]:
                        minimum = G[i][j]
                        x = i
                        y = j
    print(str(x) + "-" + str(y) + ":" + str(G[x][y]))
    selected_vertex[y] = True
    no_of_edges += 1
```

```python
import networkx as nx 
import matplotlib.pyplot as plt
G = nx.Graph() 
edges=[('a','b',1),('a','d',4),('d','c',3)]
G.add_weighted_edges_from(edges) 
pos=nx.spring_layout(G) # pos = nx.nx_agraph.graphviz_layout(G)
nx.draw_networkx(G,pos)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
plt.show()        

```
