```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head

    def display(self):
        if not self.head:
            return
        temp = self.head
        print(temp)
        while True:
            print(temp.data, end=' -> ')
            temp = temp.next
            if temp == self.head:
                break
        print()
    def AtBegining(self, data):  
        #Create a new node  
        newNode = Node(data) 
          
        #Checks if the list is empty  
        if(self.head == None):  
            #If list is empty, both head and tail will point to new node  
            self.head = newNode 
             
        else:  
            #Node temp will point to head  
            temp = self.head;  
            #newNode will become new head of the list  
            self.head = newNode;  
            #Node temp(previous head) will be added after new head  
            self.head.next = temp;   

    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head
    def insert_at_position(self, data, position):
            new_node = Node(data)
            if position == 0:
                new_node.next = self.head
                self.head = new_node
                temp = self.head
                while temp.next != self.head:
                    temp = temp.next
                temp.next = new_node
            else:
                temp = self.head
                for _ in range(position - 1):
                    temp = temp.next
                new_node.next = temp.next
                temp.next = new_node

# Example usage:
circular_list = CircularLinkedList()
circular_list.append(1)
circular_list.append(2)
circular_list.append(3)

print("Original circular linked list:")
circular_list.display()

circular_list.add_at_end(4)  # Adding 4 at the end
circular_list.insert_at_position(11,2)
print("\nCircular linked list after adding element at the end:")
circular_list.display()
```
